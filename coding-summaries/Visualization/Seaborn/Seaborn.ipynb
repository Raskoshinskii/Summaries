### Seaborn
Seaborn is an open source, BSD-licensed Python library providing high level API for visualizing the data using Python programming language. It is built on top of Matplotlib library.
### Seaborn DataSets
- **sns.get_dataset_names( )** - get all available datasets
- **sns.load_dataset( ' name ' )** - loads provided dataset 

### Seaborn Figure Styles
- **sns.set ( 'style' )** - sets provided style ( Darkgrid, Whitegrid, Dark, White, Ticks )
- **sns.despine( )** - removes top and right axes
- **sns.axes_style( )** - returns current set of styles. Can be customized by passing a new **dictionary** to the **set_style( )** 
- **sns.set_style( 'style', { new_set_of_parameters } )** - customizes current style by passing a dictionary

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline 
```

### Histogram
Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.
- **sns.distplot( df, bins, hist, kde )** - plots data distribution  

Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable. It is used for non-parametric analysis.

Setting the hist flag to False in distplot will yield the kernel density estimation plot.

```python
data = sns.load_dataset('iris')
sns.set_style('ticks')
plt.grid(True)
sns.distplot(data['petal_length'],kde=True,hist=True)
plt.show()
```

### Bivariate Distribution
Bivariate Distribution is used to determine the relation between two variables. This mainly deals with relationship between two variables and how one variable is behaving with respect to the other.
- **sns.jointplot( )** - plots relationships between several variables
- **sns.pairplot( )** - represents pairwise relationship between features

```python
# Joint pot (Scatter)
sns.set_style('darkgrid')
sns.jointplot(data.sepal_length,data.petal_length)
plt.show()
```

### Hexbin plot
Hexagonal binning is used in bivariate data analysis when the data is sparse in density i.e., when the data is very scattered and difficult to analyze through scatterplots.

```python
sns.jointplot(x='petal_length',y='petal_width',data=data,kind='hex')
plt.show()
```

### Kernel Density Estimation
Kernel density estimation is a non-parametric way to estimate the distribution of a variable. In seaborn, we can plot a kde using jointplot().

```python
sns.jointplot(x = 'petal_length',y = 'petal_width',data = data,kind = 'kde')
plt.show()
```

### Visualizing Pairwise Relationship
To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. This shows the relationship for (n,2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.

```python
sns.pairplot(data,hue='species',diag_kind='kde',kind='scatter',palette='husl')
plt.show()
```

### Strip Plot
When one or both the variables under study are categorical, we use plots like striplot(), swarmplot(). It represents the data in sorted order along any one of the axis.
- **sns.stripplot( )**

```python
sns.stripplot(x='species',y='petal_length',data=data,jitter=True)
plt.show()
```

### Swarm Plot
This function positions each point of scatter plot on the categorical axis and thereby avoids overlapping points.
- **sns.swarmplot( )**

```python
sns.swarmplot(x='species',y='petal_length',data=data)
plt.show()
```

### Box Plots
Boxplot is a convenient way to visualize the distribution of data through their quartiles. 

In **categorical scatter plots** the approach becomes limited in the information it can provide about **the distribution of values in each category.**
- **sns.boxplot( )**

```python
sns.set_style('darkgrid')
sns.boxplot('species','sepal_length',data=data)
plt.show()
```

```python
sns.boxplot(data=data,orient='h');
```

### Violin Plots
Violin Plots are a combination of the box plot with the kernel density estimates.

The quartile and whisker values from the boxplot are shown inside the violin. As the violin plot uses KDE, the wider portion of violin indicates the higher density and narrow region represents relatively lower density. The Inter-Quartile range in boxplot and higher density portion in kde fall in the same region of each category of violin plot.
- **sns.violinplot( )**

```python
data = sns.load_dataset('tips')
sns.violinplot('day','total_bill',data=data)
plt.show()
```

```python
# Taking Sex into account 
sns.violinplot('day','total_bill',hue='sex',data=data)
plt.show()
```

```python
# Provide split = True
sns.violinplot('day','total_bill',hue='sex',split=True,data=data)
plt.show()
```

### Bar Plot
The barplot() shows the relation between a categorical variable and a continuous variable. The data is represented in rectangular bars where the length the bar represents the proportion of the data in that category.

Bar plot represents the estimate of central tendency.
- **sns.barplot( )**

```python
data = sns.load_dataset('titanic')
sns.barplot('sex','survived',data=data,hue='class')
plt.show()
```

### Count Plot
To show number of observations in each category
- **sns.countplot( )**

```python
sns.countplot('class',data=data)
plt.show()
```

### Point Plot
Point plots serve same as bar plots but in a different style. Rather than the full bar, the value of the estimate is represented by the point at a certain height on the other axis.
- **sns.pointplot( )**

```python
sns.pointplot(data.sex,data.survived,hue=data.embarked);
```

### Factor Plot
Factorplot draws a categorical plot on a FacetGrid. Using âkindâ parameter we can choose the plot like boxplot, violinplot, barplot and stripplot. FacetGrid uses pointplot by default.

In Factor Plot, data is plotted on a facet grid. **Facet grid** forms a matrix of panels defined by row and column by dividing the variables. Due of panels, a single plot looks like multiple plots. It is very **helpful to analyze all combinations** in two discrete variables.
- **sns.factorplot( )**

```python
data = sns.load_dataset('exercise')
sns.factorplot(x='time',y='pulse',hue='kind', data=data);
```

```python
sns.factorplot(x='time',y='pulse',hue='kind',kind = 'box',data=data);
```

```python
sns.factorplot(x='time',y='pulse',hue='kind',kind='point',col='diet',data=data);
```

```python
data = sns.load_dataset('titanic')
sns.factorplot('alive',col='deck',data=data[data.deck.notnull()],kind='count',col_wrap = 3);
```

### Linear Relationships
There are two main functions in Seaborn to visualize a linear relationship determined through regression. These functions are regplot() and lmplot().
- **regplot( )** - accepts the x and y variables in a variety of formats including simple numpy arrays, pandas Series objects, or as references to variables in a pandas DataFrame.
- **lmplot( )** - has data as a required parameter and the x and y variables must be specified as strings. This data format is called âlong-formâ data

**We can also fit a linear regression when one of the variables takes discrete values**

The simple linear regression model used below is very simple to fit, but in most of the cases, the data is non-linear and the below methods cannot generalize the regression line

```python
data = sns.load_dataset('tips')
sns.set_style('whitegrid')
sns.regplot('total_bill','tip',data=data,color='purple');
```

```python
sns.lmplot('total_bill','tip',data=data);
```

```python
# Discrete variables
sns.lmplot('size','tip',data=data);
```

```python
# It works perfectly 
data = sns.load_dataset('anscombe')
sns.lmplot('x','y',data=data.query("dataset == 'I'")); 
```

```python
# Another case
sns.lmplot('x','y',data=data.query("dataset =='II'"));
# The relationship is not linear!
```

```python
sns.lmplot('x','y',data=data.query("dataset =='II'"),order=2);
```

### Facet Grid
A useful approach to explore medium-dimensional data, is by drawing multiple instances of the same plot on different subsets of your dataset. This technique is commonly called as âlatticeâ, or âtrellisâ plotting, and it is related to the idea of âsmall multiplesâ.

FacetGrid object takes a dataframe as input and the names of the variables that will form the row, column, or hue dimensions of the grid.

The variables should be categorical and the data at each level of the variable will be used for a facet along that axis.
- **g = sns.FacetGrid( )**

```python
data = sns.load_dataset('tips')
g = sns.FacetGrid(data,col='time',hue='smoker')
g.map(plt.scatter,'tip','total_bill')
g.add_legend();
```

```python
g = sns.FacetGrid(data,col='time',hue='smoker')
g.map(plt.hist,'tip');
```

### Pair Grid
PairGrid allows us to draw a grid of subplots using the same plot type to visualize data.

Unlike FacetGrid, it uses different pair of variable for each subplot. It forms a matrix of sub-plots. It is also sometimes called as âscatterplot matrixâ
- **g = sns.PairGrid( )**

```python
data = sns.load_dataset('iris')
g = sns.PairGrid(data,hue='species')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter);
```

```python
# More flexible 
g = sns.PairGrid(data,hue='species')
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot,shade=True)
g.map_lower(plt.scatter);
```
