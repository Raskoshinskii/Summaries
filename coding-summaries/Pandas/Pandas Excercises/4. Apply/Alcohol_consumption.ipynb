### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
# My solution
def maturity(age_column):
    age_encoding = lambda age: 1 if age > 17 else 0
    return age_column.apply(age_encoding)

data['legal_drinker'] = maturity(data['age'])
data.head()
```

```python
# Author's solution 
def maturity(val):
    if val > 17:
        return True
    else:
        return False
    
data['legal_drinker_bool'] = data['age'].apply(maturity)
data.head()
```
