### Pandas Visualisation
**Pandas** itself cannot visualise the data. It uses another library which is a part of ecosystem ( **Matplotlib/ Seaborn** )
- **pd.plot( )** - just make a plot of a Series or Data Frame object<br>
- **pd.plot( kind = ' bar ' )** makes a bar plot
- **pd.hist()** - makes a histogram
- **pd.boxplot()** - makes a boxplot
- **scatter_matrix( )** - creates a Scatter Matrix. It is very helpful to **identify linear correlation between features**

The plot itself can be changed by using matplotlib
 - **plt.title(' Name ')** - defines a title of a plot
 - **plt.x/ylabel(' Name ' )** - defines x/y labels of aplot
 
Configuration of plots can be any. For more info chech the documentation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

### Plot

```python
ts = pd.Series(np.random.randn(1096),index=pd.date_range('2012/01/01','2014/12/31'))
ts_walk = ts.cumsum() # Series
df_walk = pd.DataFrame(ts_walk) # Data Frame 
df_walk.plot();
```

### Bar Plot

```python
df = pd.DataFrame(np.random.randn(10,4),columns=list('abcd'))
df.plot(kind='bar',stacked=True);
```

```python
df.plot(kind='barh',stacked = True);
```

### Histograms

```python
df = pd.DataFrame(np.random.randn(1000,4),columns=list('abcd'))
df.hist(bins = 100);
```

### Box Plot

```python
df = pd.DataFrame(np.random.randn(10,5))
df.boxplot(return_type='axes');
```

### Scatter Plot

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=list('ab'))
df.plot(kind='scatter',x='a',y='b');
```

### Scatter Matrix


```python
from pandas.plotting import scatter_matrix

df = pd.DataFrame(np.random.randn(1000,4),columns=list('abcd'))
scatter_matrix(df,alpha=0.2,figsize=(6,6),diagonal='kde');
```
