### A/B Testing

An A/B test, also known as a split test, is a method used in marketing and experimentation to compare two versions of a webpage, app, or other digital content. We can test absoluttely anything from text size/color to machine learning algorithms.The goal is to determine which version performs better in terms of user engagement, conversion rates, or other desired outcomes/metrics.

In an A/B test, two variants – A and B – are presented to different groups of users, with only one element differing between them. By analyzing the results, businesses can make data-driven decisions to optimize their content and achieve better results.

It is possible to test several options at the same time. However, it makes the A/B test design more complex and requires more observations in each group. Besides, we need to use "multiple comparison" between the groups. For example, if we have 3 options of a registration form we need to compare the results in this way: A/B, B/C, C/A and only then make a final decision.



### Main Steps 
1. Formulate a hypothesis
2. Define the main KPI/metric
3. Define sample size and time frame for the test
4. Test launching, Results analysis

### 1. Hypothesis Formulation 
Any testing hypothesis must be precisely formulated.
- Correct option: Visual B increase CTR by 5% in comparison with visual A
- Incorrect opyion: Visual B is better than visual A

It's **unclear what word "better" means. Thus, we have to use metrics to quantify the improvement.** Probably, minor improvements are not interested (0.001 change in CTR/Conversion Rate) and there is no need for such experiments, though B in this case will be "better" than A.

For example, let's assume that we test a new RecSys algorithm and we measure Conversion Rate and CTR. However, he comes in another question - "What metric is better?"

### 2. What Metric to Choose for the Test?
If there are no market requirements it might be a bit tricky. It's recommended to **choose a metric that is super close to money - Conversion Rate.** However, due to **"sales funnel"** we will have less users/little data for that metric. It means that we will need more users and more time for the experiment to test it properly. 

For this purpose we need to **have alternative metrics that are correlated with the target metric (e.g. Conversion Rate).** As an alternative we can use CTR. With this metric we will need less users and less time for the test and can test a feature much faster.

Altough we have only one main metric in the test, we can have additional ones. It may happen that some two metrics may show absolutely different results in two groups. For example:
- Group A: CTR - 4% | CR - 5%
- Group B: CTR - 10% | CR - 2%

In this case we must revisit A/B test design. Probably, we show recommendation to users that like it but not targeted (not ready for a purchase).


### 3. Sample Size Calculation
We need to clearly understand when we can finish A/B test. Alternatively, we can say a random time period - in a week or two weeks. However, what if even two weeks is not enough? To avoid such problems we need to calculate a sample size for the test to have enough samples to say with a certain confidence that we can stop the test. 

**Why Do We Need It?**

**Proper sample size allows avoid "peeping" at A/B test results.**



We can peep in the results to make sure that everything is working smoothly/as expected.

**We cannot finsih the test early or late when peeping.** We must always stopt the test when we have enough samples to make a conclusion. 

For example, we may have enough samples for stat test, we test and see that we almost there, just need 3-10 more observations till significant results. If we do like this, it may lead to false conclusion ("peeping trap"). 

**Main Parameters for Sample Calculation**

- Main testing KPI: CTR, Conversion Rate
- Reliability: 90/95% (tells how confident we are if there is significant difference)
- Test Power: 80% (tells how confident we are if there is no difference between two options)
- Expected Effect Size: 5/10/15% uplift

### 4. Results Analysis
After reaching the required sample size test must be "switched off". **Gathered data is tested using a statistical test:**
- Chi Squared Test
- Fisher's Exact Test
- T-test
- ...

It's a good practive to validate that users were independent in the groups.

### A/A testing 
Since groups must represent independent users we need to have a method that will allow cheking it. In A/A testing we test users with no improvent to see if there is difference between them. We know that should be no differecne but if the test shows that such difference exist it gives a hint that two groups are dependent and we need to resplit them randomly.

We can use AAB testing: we split users into 3 groups and check if groups are independent and if there is improvement in B.

### Important Notes
- **Don't test big changes at once:** if difference is discovered, it's hard to say what was the reason for that: text, color, placement
- **Minor effect size requires more observations in groups**: revisit Power Analysis topic
