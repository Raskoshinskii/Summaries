### 2. try - except - else - finally

Данные блоки команд служат для отслеживания и отладки исключений, возникаемые в ходе выполениея программы

- `try` - помещается код, где вероятны ошибки
- `except` - код, определяющий логику обработки возникающих исключений
- `else` - код, который выполнится если не возникли исключения
- `finaly` - код, который выполнится всегда, даже если воникли исключения 



```python
# определим фунцию деления 2-х чисел и список
logs = [] # хранит историю значений передаваемых аргументов и результат функции div_values
def div_values(a, b, logs=logs):
    try:
        res = a / b
        print('\nРезультат деление: ', res)
        logs.append((a, b, "успех"))
    except ZeroDivisionError:
        print('\nДеление на ноль, используй другое число')
        logs.append((a, b, "деление на ноль"))
    else:
        print('Деление успешно!')
    finally:
        print(f'Первый аргумент: {a}, второй: {b}')
```

```python
div_values(2, 4)
div_values(10, 2)
div_values(10, 0)
```
