```python
# Necessary Libraries
import requests
import fake_useragent
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm as log_progress
```

### No Multiprocessing

```python
home_page = 'https://coinmarketcap.com/'
page_start = 1
page_end = 42
coin_names = []

# Iteration Over All Pages
for page in log_progress(range(page_start, page_end)):
    current_url = f'{home_page}{page}/'
    
    # Data for Headers (Update User-Agent For Every New Page)
    user = fake_useragent.UserAgent().random
    header = {'user-agent': user}

    # HTML Getting
    response = requests.get(current_url, headers = header).text
    html = BeautifulSoup(response, 'lxml')

    # Finding Coin Names
    target_table = html.find('table', class_ = 'cmc-table cmc-table___11lFC cmc-table-homepage___2_guh')
    try:
        target_table = target_table.find('tbody')
        table_rows = target_table.find_all('tr')
    except:
        print('Target Table Not Found')

    # Coin Names Extraction
    for table_row in table_rows:
        try:
            coin_name = table_row.find('a').get('href').split('/')[2]
            print(f'{coin_name} has been extracted')
            coin_names.append(coin_name)
        except:
            print('Coin Is Not Extracted')
```
