### Matplotlib
Matplotlib is one of the most popular Python packages used for data visualization. It is a cross-platform library for making 2D plots from data in arrays. 

It provides an object-oriented API that helps in embedding plots in applications using Python GUI toolkits such as PyQt, WxPythonotTkinter. It can be used in Python and IPython shells, Jupyter notebook and web application servers also.

To display plot outputs inside the notebook itself (and not in the separate viewer), enter the following magic statement:

%matplotlib inline


### Plot Customization Methods
rcParams stands for Run/Configur

You can also dynamically change the default rc settings in a python script or interactively from the python shell. All of the rc settings are stored in a dictionary-like variable called matplotlib.rcParams, which is global to the matplotlib package. rcParams can be modified directly:
- **plt.rcParams[' axes.labelsize ']** - sets label size
- **plt.rcParams[' xtick.labelsize ']** - sets xticks size
- **plt.rcParams[' ytick.labelsize ']** - sets ytick size


- **plt.x/ylabel(' x/y labels ')** - labels x/y axes
- **plt.title( ' Titel ' )** - sets a title
- **plt.grid( True )** - sets grid. **It can be customized as well**
- **plt.subplot( nrows, ncols, index )** - creates subplots with provided dimension.
- **plt.tight_layout( )** - to prevent overlapping between axes

### Figure Customization Methods
The matplotlib.figure module contains the Figure class. It is a top-level container for all plot elements.  

**A given figure can contain many Axes, but a given Axes object can only be in one Figure**

Axes object is added to figure by calling the add_axes() method. It returns the axes object and adds an axes at position rect [left, bottom, width, height] where all quantities are in fractions of figure width and height
- **figsize ( )** - the size of a plot 
- **dpi( )** - plot quality
- **fig.add_axes( [ initial_coordinates ] )** - creates a new **axes object**

### Axes Customization Methods
- **ax.set_title(' Title ' )** - sets a title
- **ax.set_x/ylabel( 'x/y labels ' )** - labels x/y axes
- **ax.grid(True)** - sets grid. **It can be customized**
- **ax.plot( )** - plots current axes
- **ax.legend( )** - plots provided legend 
- **ax.set_x/yscale( )** - sets provided scale on the axis
- **ax.x/yaxis.labelpad = val** - adds apdding from x/y labels to the spine
- **ax.spines[ ' position ' ].set_color( ' color ' )** - changes the color of the axis
- **ax.spines[ ' position ' ].set_linewidth( )** - changes the size of the provided spine
- **ax.set_x/ylim( start, end )** - changes x/y axes limits
- **ax.set_x/yticks( [ values ] )** - creates provided ticks on x/y axes
- **ax.set_x/yticklabels( [ ' labels ' ] )** - creates labels for the ticks 
- **ax.twinx/y( )** - Create a new Axes instance with an invisible x-axis and an independent y-axis positioned opposite to the original one (i.e. at right).

### Subplots
- **plt.subplot( nrows, ncols, index )** - creates subplots with provided dimension.
- **ax = fig.add_subplot( intial coordinates )** - creates several subplots
- **fig, ax = plt.subplots ( nrows, ncols )** - creates subplots according to provided dimension
- **ax = plt.subplot2grid(shape, loc, rowspan, colspan)**

### Bar Plot
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value.
- **ax.bar( )** 

### Histogram 
A histogram is an accurate representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable.

To construct a histogram, follow these steps:
- Bin the range of values.
- Divide the entire range of values into a series of intervals.
- Count how many values fall into each interval.

The bins are usually specified as consecutive, non-overlapping intervals of a variable
- **ax.hist( )**

### Pie Chart
A Pie Chart can only display one series of data. Pie charts show the size of items (called wedge) in one data series, proportional to the sum of the items. The data points in a pie chart are shown as a percentage of the whole pie.
- **ax.pie( data, labels, autopct = %1.2f%%)** 

### Scatter Plot
Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to show how much one variable is affected by another. Each row in the data table is represented by a marker the position depends on its values in the columns set on the X and Y axes. A third variable can be set to correspond to the color or size of the markers, thus adding yet another dimension to the plot.
- **ax.scatter( )**

### Contour Plot 
Contour plots (sometimes called Level Plots) are a way to show a three-dimensional surface on a two-dimensional plane. It graphs two predictor variables X Y on the y-axis and a response variable Z as contours. These contours are sometimes called the z-slices or the iso-response values
- **ax.contourf( )** - filled gaph
- **ax.contour( )** - not filled graph

### Box Plot
A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum
- **ax.boxplot( )** 

### Violin Plot 
Violin plots are similar to box plots, except that they also show the probability density of the data at different values.
- **ax.violinplot( )** 

### 3D Plotting 
Even though Matplotlib was initially designed with only two-dimensional plotting in mind, some three-dimensional plotting utilities were built on top of Matplotlib's two-dimensional display in later versions, to provide a set of tools for three-dimensional data visualization. 

Three-dimensional plots are enabled by importing the mplot3d toolkit, included with the Matplotlib package.
- **ax = plt.axes( projection = '3d' ); ax.plot3D( )**
- **ax.scatter(x, y, z)** - plots 3D scatter plot
- **ax.surface_plot( )** - plots 3D surface

### Simple Plot 

```python
x = np.arange(0,math.pi*2,0.05)
y = np.sin(x)

plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine')
plt.title('Sine Plot');
plt.grid(True)
```

### Plot Using Figure and Axes Objects 

```python
# Using object - oriented interface. Provides more fexability becasue we don't need to create the same figure several times.
# Instead, we can create as many different plots as we want from the same figure object!

fig = plt.figure(figsize=(3,2), dpi=150) # creates an empty canvas which can be customized
ax = fig.add_axes([0,0,1,1]) # add axes to the figure ( it sets the dimension of axes with initial coordiantes (0;0),(1,1) )
ax.plot(x,y)
ax.set_title('Sine Plot')
ax.set_xlabel('angle')
ax.set_ylabel('sine')
ax.grid(True)
```

```python
# Several Lines
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]

fig = plt.figure(figsize=(3,2), dpi=150)
ax = fig.add_axes([0,0,1,1])

line_1 = ax.plot(x1,y,'ys-')
line_2 = ax.plot(x2,y,'go--')

ax.legend(labels = ('tv','phones'),loc='lower right')
ax.set_title("Advertisement effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```

```python
# Using Figure object
x = np.arange(0,math.pi*2,0.05)
y_1 = np.sin(x)
y_2 = np.cos(x)

fig = plt.figure(figsize=(3,2),dpi=150)
ax_1 = fig.add_axes([0,0,1,1])
ax_2 = fig.add_axes([0.15,0.2,0.3,0.3])

ax_1.plot(x,y_1,'b')
ax_2.plot(x,y_2,'r')

ax_1.set_title('sine')
ax_2.set_title('cosine')
plt.show()
```

### Subplots( )

```python
fig, ax = plt.subplots(2,2)

# First Subplot
x = np.arange(1,5)
ax[0][0].plot(x,x**2)
ax[0][0].set_title('Squared')

# Second Subplot
ax[0][1].plot(x,np.sqrt(x))
ax[0][1].set_title('Square Root')

# Third Subplot
ax[1][0].plot(x,np.exp(x))
ax[1][0].set_title('Exp')

# Forth Subplot
ax[1][1].plot(x,np.log10(x))
ax[1][1].set_title('Log')

plt.tight_layout()
plt.show()
```

### Subplot2Grid( )

```python
# Grid Initialization 
ax_1 = plt.subplot2grid((3,3),(0,0),colspan=2)
ax_2 = plt.subplot2grid((3,3),(0,2),rowspan=3)
ax_3 = plt.subplot2grid((3,3),(1,0),colspan=2,rowspan=2)

x = np.arange(1,10)

# Second Axes Subpot
ax_2.plot(x,x**2)
ax_2.set_title('Square')

# First Axes Subplot 
ax_1.plot(x,np.exp(x))
ax_1.set_title('Exp')

# Third Axes
ax_3.plot(x,np.log10(x))
ax_3.set_title('Log')

plt.tight_layout()
plt.show()
```

### Grids
The grid() function of axes object sets visibility of grid inside the figure to on or off. You can also display major / minor (or both) ticks of the grid. Additionally color, linestyle and linewidth properties can be set in the grid() function

```python
fig, ax  = plt.subplots(1,3, figsize=(12,4) )

x = np.arange(1,11)

# First Grid 
ax[0].plot(x,x**3,'g',lw=1)
ax[0].grid(True)
ax[0].set_title('Default Grid')

# Second Grid 
ax[1].plot(x,np.exp(x),'r',lw=1)
ax[1].grid(color='b',ls='-.',lw=0.3)
ax[1].set_title('Custom Grid')

# Third Grid 
ax[2].plot(x,x)
ax[2].set_title('No Grid')

plt.tight_layout()
plt.show()
```

### Axes Formatting

```python
fix, ax = plt.subplots(1,2, figsize = (10,4))

x = np.arange(1,5)

# Plotting 
ax[0].plot(x,np.exp(x))
ax[0].plot(x,x**2)
ax[1].plot(x,np.exp(x))
ax[1].plot(x,x**2)


# Scaling 
ax[1].set_yscale('log')
ax[1].xaxis.labelpad = 10
ax[0].xaxis.labelpad = 10 # the distance from xlabel to the spine

# Titles
ax[0].set_title('Normal Scale')
ax[1].set_title('Log Scale')

# Grids 
ax[0].grid(True)
ax[1].grid(True)

# Axes Labels
ax[0].set_xlabel('x axis')
ax[0].set_ylabel('y axis')
ax[1].set_xlabel('x axis')
ax[1].set_ylabel('y axis')

#####################
# Spines Formatting #
#####################

# 1.1 Changing colors
ax[0].spines['bottom'].set_color('red')
ax[0].spines['top'].set_color('red')
ax[0].spines['left'].set_color('red')
ax[0].spines['right'].set_color('red')

# 1.2 Changing line width
ax[0].spines['bottom'].set_linewidth(2)
ax[0].spines['top'].set_linewidth(2)
ax[0].spines['left'].set_linewidth(2)
ax[0].spines['right'].set_linewidth(2)

# 1.3 Changing spines limits
ax[0].set_ylim(0,10)
ax[0].set_xlim(1,2)

plt.tight_layout()
plt.show()
```

### Ticks and Ticks Labels

The xticks() and yticks() function takes a list object as argument. The elements in the list denote the positions on corresponding action where ticks will be displayed.

```python
# Sine Creation
x = np.arange(0,math.pi*2,0.05)
y = np.sin(x)

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])

# Let's change the plot and its ticks
ax.set_xlabel('angle')
ax.set_ylabel('value')
ax.set_title('Sine Plot')
ax.set_xticks([0,2,4,6])
ax.set_xticklabels(['zero','two','four','six'])
ax.set_yticks([-1,0,1])
ax.plot(x,y)
plt.show()
```

### Dual Axes

It is considered useful to have dual x or y axes in a figure. Moreover, when plotting curves with different units together. 

Matplotlib supports this with the twinxand twiny functions.

```python
fig = plt.figure()
ax_1 = fig.add_axes([0,0,1,1])

x = np.arange(1,11)
y_1 = np.exp(x)
y_2 = np.log10(x)

ax_1.plot(x,y_1)
ax_1.set_ylabel('Exp')

ax_2 = ax_1.twinx()

ax_2.plot(x,y_2,'ro-')
ax_2.set_ylabel('Log')

fig.legend(labels = ('exp','log'),loc='upper left')
plt.show()
```

### Bar Plot

```python
# Simple Bar Plot 
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
prog_lang = ['Python','C','C++','PHP','Java']
students = [23,17,10,19,20]
ax.bar(prog_lang,students)
plt.show()
```

```python
# Several Bar Plots 
data = [[30,25,50,20],
        [40,23,51,17],
        [35,22,45,19]]
x = np.arange(4)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(x + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(x + 0.5, data[2], color = 'r', width = 0.25)
plt.show()
```

```python
# Stacked Bar Plot 

# Data
men = (20, 35, 30, 35, 27)
women = (25, 32, 34, 20, 25)

# Parameters
idx = np.arange(len(men)) # location for the groups on x axis 
width = 0.35

# Plotting
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(idx, men, width,color='g')
ax.bar(idx, women, width, bottom = men,color='b')

# Labeling
ax.set_title('Scores by group and gender')
ax.set_ylabel('Scores')
ax.legend(labels=['Men','Women'])
plt.show()
```

### Contour Plot

```python
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)

x,y = np.meshgrid(x,y)
z = np.sqrt(x**2 + y**2)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
graph = ax.contourf(x,y,z)
fig.colorbar(graph)
ax.set_title('Contour Plot of Paraboloid')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

### 3D Plotting 

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0,1,100)
x = z*np.sin(20*z)
y = z*np.cos(20*z)
ax.plot3D(x,y,z,'gray')
ax.set_title('3D Plot')
plt.tight_layout()
plt.show()
```

### Surface Plot

```python
x = np.outer(np.linspace(-2,2,30),np.ones(30))
y = x.copy().T # transpose
z = np.cos(x**2 + y**2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z,cmap='viridis',edgecolor='none')
ax.set_title('Surface Plot')
plt.show()
```
